```python
# To support both python 2 and python 3
from __future__ import division, print_function, unicode_literals

# Common imports
import numpy as np

# to make this notebook's output stable across runs
np.random.seed(42)

# To plot pretty figures
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

import pandas as pd

from sklearn.linear_model import SGDClassifier
```

# Load training classifier
# Set row and column positions to variables X and Y

```python
mnist = pd.read_csv("train.csv")
X = mnist.iloc[0:,1:]
Y = mnist.iloc[0:,0]
```

# Get a position values to test train set

```python
X_train = X.iloc[:4324, :]
Y_train = Y.iloc[:4324]
```

```python
Y_train_Num = (Y_train == 1)
```

# Create a SGD classifer and train model with test position
# Try and predict at particular X positions

```python
sgdClass = SGDClassifier(random_state = 42)
sgdClass.fit(X_train, Y_train_Num)
sgdClass.predict([X.iloc[8,:]])
```

```python
sgdClass.predict([X.iloc[4324,:]])
```

# Use SGD to train the classifier using the whole training set

```python
sgdClass.fit(X, Y)
```

# Load test set and predict positions based on trained classifier

```python
tmnist = pd.read_csv("test.csv")
```

```python
classifier = sgdClass.predict(tmnist)
classifier
```

# Put the classifier data into data set and then output it to file

```python
indexes = range(1, classifier.size + 1)
class_data = pd.DataFrame({
    'ImageId': indexes,
    'Label': classifier
})

class_data[['ImageId', 'Label']].to_csv('sgd_class.csv', index = False, encoding='utf-8')
```
